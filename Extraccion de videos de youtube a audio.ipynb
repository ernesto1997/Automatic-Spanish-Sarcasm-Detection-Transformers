### Extracción del audio de los videos de youtube

```python
# Función para obtener el siguiente nombre disponible para el archivo de audio
def get_next_filename(output_dir, base_name="DosHombresYMedio", output_format="wav"):
    i = 1
    while True:
        # Construir el nombre del archivo basado en el número
        audio_filename = f"{base_name}{i}.{output_format}"
        # Revisar si el archivo ya existe en el directorio
        if not os.path.exists(os.path.join(output_dir, audio_filename)):
            return audio_filename  # Retornar el primer nombre disponible
        i += 1

# Función para descargar el video de YouTube
def download_video(url):
    try:
        yt = YouTube(url)
        video_title = yt.title  # Obtener el título del video
        # Descargar el video con un nombre temporal
        video_file = yt.streams.get_highest_resolution().download(filename=f'{video_title}.mp4')
        print(f"Video descargado: {video_file}")
        return video_file, video_title
    except Exception as e:
        print(f"Error al descargar el video de la URL {url}: {e}")
        return None, None

# Función para extraer el audio del video usando moviepy
def extract_audio_with_moviepy(video_file, output_dir='audios_DosHombresYMedio', output_format='wav'):
    try:
        # Crear la carpeta "audios" si no existe
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)

        # Obtener el siguiente nombre disponible para el archivo de audio
        audio_filename = get_next_filename(output_dir, base_name="DosHombresYMedio", output_format=output_format)
        audio_file = os.path.join(output_dir, audio_filename)

        # Usar context manager para asegurar que el archivo de video se cierra después de extraer el audio
        with VideoFileClip(video_file) as video_clip:
            # Guardar el audio con el nombre basado en "House" + número
            video_clip.audio.write_audiofile(audio_file)
            print(f"Audio extraído y guardado como: {audio_file}")
        return audio_file
    except Exception as e:
        print(f"Error al extraer el audio: {e}")
        return None

# Función para procesar una lista de URLs desde un archivo
def process_videos_from_file(input_file):
    try:
        # Leer el archivo de URLs
        with open(input_file, 'r') as file:
            urls = file.readlines()

        for url in urls:
            url = url.strip()  # Eliminar espacios y saltos de línea
            if not url:
                continue  # Saltar líneas vacías
            print(f"Procesando video: {url}")

            # Descargar el video de YouTube
            video_file, video_title = download_video(url)
            if video_file is None or video_title is None:
                continue

            # Extraer el audio del video descargado
            extract_audio_with_moviepy(video_file)

            # Intentar eliminar el archivo de video después de extraer el audio
            if os.path.exists(video_file):
                try:
                    os.remove(video_file)
                    print(f"Archivo de video eliminado: {video_file}")
                except Exception as e:
                    print(f"No se pudo eliminar el archivo de video: {e}")

    except Exception as e:
        print(f"Error al procesar los videos: {e}")

# Archivo que contiene la lista de URLs de YouTube
input_file = 'urls_videos_DosHombresYMedio.txt'  # Cambia esto por el nombre de tu archivo de URLs

# Procesar los videos desde la lista
process_videos_from_file(input_file)
```
