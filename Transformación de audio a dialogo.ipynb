```python
import assemblyai as aai
import csv
import os

aai.settings.api_key = "eb992327c14744a68adc975f5bfb936a"

# Función para obtener todos los archivos de audio desde una carpeta
def get_audio_files_from_folder(folder_path, extension=".wav"):
    audio_files = []
    try:
        # Recorrer todos los archivos en la carpeta
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                if file.endswith(extension):
                    # Agregar archivos con la extensión especificada
                    audio_files.append(os.path.join(root, file))
        return audio_files
    except Exception as e:
        print(f"Error al obtener archivos de audio de la carpeta: {e}")
        return []

# Ruta a la carpeta que contiene los archivos de audio
audio_folder_path = r"C:\Users\User\TFM\Dialogos\audios_DosHombresYMedio" 

# Obtener todos los archivos de audio de la carpeta
audio_files = get_audio_files_from_folder(audio_folder_path)

# Configuración de la transcripción
config = aai.TranscriptionConfig(
    speaker_labels=True,   # Detección de interlocutores
    language_code="es"     # Especificar español como idioma
)

# Crear transcriptor
transcriber = aai.Transcriber()

# Crear o abrir archivo CSV con separador ';'
with open("DosHombresYMedio_sarcasmo4.csv", mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file, delimiter=';')
    
    # Escribir los encabezados
    writer.writerow(["Frase Anterior", "Frase Actual"])

    # Procesar cada archivo de audio
    for audio_file in audio_files:
        if not os.path.exists(audio_file):
            print(f"El archivo no existe: {audio_file}")
            continue
        
        print(f"Procesando archivo: {audio_file}")
        
        # Transcribir el archivo de audio
        transcript = transcriber.transcribe(
            audio_file,
            config=config
        )
        
        # Variable para almacenar la frase anterior
        frase_anterior = None
        
        # Procesar cada transcripción
        for utterance in transcript.utterances:
            frase_actual = utterance.text
            writer.writerow([frase_anterior, frase_actual])
            
            # Actualizar la frase anterior
            frase_anterior = frase_actual
```
